```python
from time import time

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy import optimize
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier

%matplotlib inline
```

```python
class MyDecisionTreeClassifier:
    NON_LEAF_TYPE = 0
    LEAF_TYPE = 1

    def __init__(self, min_samples_split=2, max_depth=None, sufficient_share=1.0, criterion='gini', max_features=None):
        self.tree = dict()
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.sufficient_share = sufficient_share
        self.num_class = -1
        if criterion == 'gini':
            self.G_function = self.__gini
        elif criterion == 'entropy':
            self.G_function = self.__entropy
        elif criterion == 'misclass':
            self.G_function = self.__misclass
        else:
            print 'invalid criterion name'
            raise

        if max_features == 'sqrt':
            self.get_feature_ids = self.__get_feature_ids_sqrt
        elif max_features == 'log2':
            self.get_feature_ids = self.__get_feature_ids_log2
        elif max_features == None:
            self.get_feature_ids = self.__get_feature_ids_N
        else:
            print 'invalid max_features name'
            raise

    def __gini(self, l_c, l_s, r_c, r_s):
        # Ваш код
        # Вычисляет -(|S_l|*I(S_L) + |S_R|*I(S_R)) где I - неопределенность по gini
        l_s = l_s.astype('float')
        r_s = r_s.astype('float')
        
        # Доли классов
        l_p = l_c / l_s
        r_p = r_c / r_s

        # Gini
        l_e = (1 - np.sum(l_p ** 2, axis=1)).reshape(l_s.shape[0], 1)
        r_e = (1 - np.sum(r_p ** 2, axis=1)).reshape(r_s.shape[0], 1)
           
        return (-(l_s * l_e + r_s * r_e)).reshape(l_s.shape[0])
    
    def __entropy(self, l_c, l_s, r_c, r_s):
        # Ваш код
        # Вычисляет -(|S_l|*I(S_L) + |S_R|*I(S_R)) где I - неопределенность по entropy
        l_s = l_s.astype('float')
        r_s = r_s.astype('float')
        
        # Доли классов
        l_p = l_c / l_s
        r_p = r_c / r_s
        
        # Entropy
        l_e = ( - np.sum(np.nan_to_num(l_p * np.log(l_p)), axis=1)).reshape(l_s.shape[0], 1)
        r_e = ( - np.sum(np.nan_to_num(r_p * np.log(r_p)), axis=1)).reshape(l_s.shape[0], 1)
              
        return (-(l_s * l_e + r_s * r_e)).reshape(l_s.shape[0])

    def __misclass(self, l_c, l_s, r_c, r_s):
        # Ваш код
        # Вычисляет -(|S_l|*I(S_L) + |S_R|*I(S_R)) где I - неопределенность по misclass
        l_s = l_s.astype('float')
        r_s = r_s.astype('float')
        
        # Доли классов
        l_p = l_c / l_s
        r_p = r_c / r_s
        
        # Misclass
        l_g = (1 - np.max(l_p, axis=1)).reshape(l_s.shape[0], 1)
        r_g = (1 - np.max(r_p, axis=1)).reshape(r_s.shape[0], 1)
              
        return (-(l_s * l_g + r_s * r_g)).reshape(l_s.shape[0])

    def __get_feature_ids_sqrt(self, n_feature):
        feature_ids = range(n_feature)
        np.random.shuffle(feature_ids)
        # Ваш код
        return np.array(feature_ids[:int(n_features ** 0.5)])
        
    def __get_feature_ids_log2(self, n_feature):
        feature_ids = range(n_feature)
        np.random.shuffle(feature_ids)
        # Ваш код
        return np.array(feature_ids[:int(np.log(n_features))])

    def __get_feature_ids_N(self, n_feature):
        # Ваш код
        return np.arange(n_feature)
    
    def __sort_samples(self, x, y):
        sorted_idx = x.argsort()
        return x[sorted_idx], y[sorted_idx]

    def __div_samples(self, x, y, feature_id, threshold):
        left_mask = x[:, feature_id] > threshold
        right_mask = ~left_mask
        return x[left_mask], x[right_mask], y[left_mask], y[right_mask]

    def __find_threshold(self, x, y):
        # Что делает этот блок кода?
        # Сортируются в порядке возрастания значения определенного столбца данных, ищутся соответсвующие им классы
        # Считается общее число классов
        sorted_x, sorted_y = self.__sort_samples(x, y)
        class_number = np.unique(y).shape[0]
        
        # Что делает этот блок кода?
        # Отсекаются из рассмотрения минимальные и максимальные значения
        # Вычисляется индексы, где меняется класс соседнего элемента
        splitted_sorted_y = sorted_y[self.min_samples_split:-self.min_samples_split]
        r_border_ids = np.where(splitted_sorted_y[:-1] != splitted_sorted_y[1:])[0] + (self.min_samples_split + 1)
        
        if len(r_border_ids) == 0:
            return float('+inf'), None
        
        # Что делает этот блок кода?
        # Cчитается число элементов в каждом интервале посточнства класса
        # Задается матрица принадлежности интервала к конкретному классу с указанием количества элементов в интервале
        # В первую строку добаляются элементы по классам из первых элементов y, используемых в качестве образца
        eq_el_count = r_border_ids - np.append([self.min_samples_split], r_border_ids[:-1])
        one_hot_code = np.zeros((r_border_ids.shape[0], class_number))
        one_hot_code[np.arange(r_border_ids.shape[0]), sorted_y[r_border_ids - 1]] = 1
        class_increments = one_hot_code * eq_el_count.reshape(-1, 1)
        class_increments[0] = class_increments[0] + np.bincount(y[:self.min_samples_split], minlength=class_number)
        
        # Что делает этот блок кода?
        # Считается число элементов слева и внутри интервала каждого класса
        # Считается число элеметов справа от интервала каждого класса
        # Считается количество элементов слева и внутри от каждого интервала
        # Считается количество элементов справа от каждого интервала
        l_class_count = np.cumsum(class_increments, axis=0)        
        r_class_count = np.bincount(y) - l_class_count
        l_sizes = r_border_ids.reshape(l_class_count.shape[0], 1)
        r_sizes = sorted_y.shape[0] - l_sizes

        # Что делает этот блок кода?
        # Вычисляется значение функции неопределенности для каждого интервала
        # Выбирается тот интервал, значение признака которого как граница для разбиения минимизирует функцию неопределенности.
        gs = self.G_function(l_class_count, l_sizes, r_class_count, r_sizes)
        idx = np.argmin(gs)
        
        # Что делает этот блок кода?
        # ВЫбирается индекс элемента, следуюшего сразу за найденным интервалом
        # Возвращается значение функции неопределенности и граничное значение, отделяющее значение по интервалу
        left_el_id = l_sizes[idx][0]
        return gs[idx], (sorted_x[left_el_id-1] + sorted_x[left_el_id]) / 2.0

    def __fit_node(self, x, y, node_id, depth, pred_f=-1):
        # Ваш код
        # Необходимо использовать следующее:
        # self.LEAF_TYPE
        # self.NON_LEAF_TYPE

        # self.tree
        # self.max_depth
        # self.sufficient_share
        # self.min_samples_split

        # self.get_feature_ids
        # self.__find_threshold
        # self.__div_samples
        # self.__fit_node
        

        y_sort = np.sort(y)
        y_r_bor = np.where(y_sort[:-1] != y_sort[1:])[0]
        
        # Если класс единственный
        if (len(y_r_bor) == 0):
            self.tree[node_id] = tuple((self.LEAF_TYPE, int(y[0]), 0))
            return
                
        # Подсчет числа элементов и вычисление индекса класса с максимальным числом элементов       
        y_r_bor = np.append(y_r_bor, [y_sort.shape[0] - 1])
        y_class = y_r_bor - np.append([-1], y_r_bor[:-1])
        idx = np.argmax(y_class)
        
        # Ограничение на глубину
        if (depth == self.max_depth):
            self.tree[node_id] = tuple((self.LEAF_TYPE, int(y_sort[y_r_bor[idx]]), 0))
            return
        
        # Ограничение на число элементов одинакового класса в узле
        if (y_class[idx]/y.shape[0] >= self.sufficient_share):
            self.tree[node_id] = tuple((self.LEAF_TYPE, int(y_sort[y_r_bor[idx]]), 0))
            return
            
        #Основная часть
        # В gs три столбца, номер индекса, значение неопределенности, граница
        features = self.get_feature_ids(x.shape[1])
        gs = np.empty((features.shape[0], 3))
        for ind in xrange(features.shape[0]):
            gs[ind][0] = features[ind]
            gs[ind][1], gs[ind][2] = self.__find_threshold(x[:,ind], y)
        ind = np.argmin(gs[:,1], axis=0)
 
        self.tree[node_id] = tuple([self.NON_LEAF_TYPE, int(gs[ind,0]), gs[ind,2]])
        
        node = self.tree[node_id]
        x_l, x_r, y_l, y_r = self.__div_samples(x, y, node[0], node[2])
        
        if(y_l.shape[0] != 0 and y_r.shape[0] != 0):
            self.__fit_node(x_l, y_l, 2*node_id + 1,depth + 1)
            self.__fit_node(x_r, y_r, 2*node_id + 2,depth + 1)    
        else:
            # разбиение не удалось.
            # Значение в отсортированном списке не поменялось, а класс объекта поменялся
            # и получилось разбиение на два класса, в одном из которых 0 элементов
            self.tree[node_id] = tuple((self.LEAF_TYPE, int(y_sort[y_r_bor[idx]]), 0))
        

    
    def fit(self, x, y):
        self.num_class = np.unique(y).size
        self.__fit_node(x, y, 0, 0) 

    def __predict_class(self, x, node_id):
        node = self.tree[node_id]
        if node[0] == self.__class__.NON_LEAF_TYPE:
            _, feature_id, threshold = node
            if x[feature_id] > threshold:
                return self.__predict_class(x, 2 * node_id + 1)
            else:
                return self.__predict_class(x, 2 * node_id + 2)
        else:
            return node[1]

    def __predict_probs(self, x, node_id):
        node = self.tree[node_id]
        if node[0] == self.__class__.NON_LEAF_TYPE:
            _, feature_id, threshold = node
            if x[feature_id] > threshold:
                return self.__predict_probs(x, 2 * node_id + 1)
            else:
                return self.__predict_probs(x, 2 * node_id + 2)
        else:
            return node[2]
        
    def predict(self, X):
        return np.array([self.__predict_class(x, 0) for x in X])
    
    def predict_probs(self, X):
        return np.array([self.__predict_probs(x, 0) for x in X])

    def fit_predict(self, x_train, y_train, predicted_x):
        self.fit(x_train, y_train)
        return self.predict(predicted_x)
```

```python
df = pd.read_csv('./cs-training.csv', sep=',').dropna()
df.head()
```

```python
x = df.as_matrix(columns=df.columns[1:])
y = df.as_matrix(columns=df.columns[:1])
y = y.reshape(y.shape[0])
```

```python
my_clf = MyDecisionTreeClassifier(min_samples_split=2)
clf = DecisionTreeClassifier(min_samples_split=2)
```

## Проверка скорости работы

```python
t1 = time()
my_clf.fit(x, y)
t2 = time()
print(t2 - t1)

t1 = time()
clf.fit(x, y)
t2 = time()
print(t2 - t1)
```

## Проверка качества работы

```python
gkf = KFold(n_splits=5, shuffle=True)
```

```python
for train, test in gkf.split(x, y):
    X_train, y_train = x[train], y[train]
    X_test, y_test = x[test], y[test]
    my_clf.fit(X_train, y_train)
    print(accuracy_score(y_pred=my_clf.predict(X_test), y_true=y_test))
```

```python
for train, test in gkf.split(x, y):
    X_train, y_train = x[train], y[train]
    X_test, y_test = x[test], y[test]
    clf.fit(X_train, y_train)
    print(accuracy_score(y_pred=clf.predict(X_test), y_true=y_test))
```
